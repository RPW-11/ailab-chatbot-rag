```python
from unstructured.partition.pdf import partition_pdf
from unstructured.chunking.title import chunk_by_title
```

```python
file_path = './A Robust Approach for Brain Tumor Detection in Magnetic Resonance Images Using Finetuned EfficientNet [GOOD].pdf'
```

```python
elements = partition_pdf(
    filename=file_path,
    strategy="hi_res",  # Use high-resolution for better image/table detection
    infer_table_structure=True,  # Enable table extraction
    extract_images_in_pdf=True,  # Extract images
    languages=["en", "zh"],  # Specify the language
    extract_image_block_output_dir="./images/",  # Directory to save images
)
```

```python
for element in elements:
    print(f"Type: {element.category}")
    print(f"Content: {element.text}")
    print("-" * 50)
```

```python
chunks = chunk_by_title(
    elements,
    max_characters=1024,  # Max chunk size
    combine_text_under_n_chars=200,  # Merge small elements
)
```

```python
for chunk in chunks:
    print(chunk.metadata.to_dict())
    print(f"Category: {chunk.category}")
    print(f"Chunk: {chunk.text}")
    print("-" * 50)
```
